```python
from bs4 import BeautifulSoup
import requests
import sys
import re
```

```python
print('Você gostaria de saber sobre o que?')
frase = input()
words = frase.split(' ')

pattern = re.compile("(\.*)*ia$")
pattern2 = re.compile("(\.*)+r$")
pattern3 = re.compile("sobre")

palavras = []

for w in words:
    if pattern.search(w) == None and pattern2.search(w) == None and pattern3.search(w) == None:
        palavras.append(w)

tam = {}
nmax = []

for palavra in palavras:
    nmax.append(len(palavra))
    tam[palavra] = len(palavra)
    
for p in palavras:
    if tam[p] == max(nmax):
        chave = p
        
cade(chave)

print('Mais algo?')
res = input()

if res.lower() == 'yes':
    buscar(chave)
```
