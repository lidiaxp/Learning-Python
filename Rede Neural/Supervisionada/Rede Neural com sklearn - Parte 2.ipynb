```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

print(cancer.feature_names)
print('-----------------------')
print(cancer.target_names)
print('-----------------------')
print(len(cancer.data))
print('-----------------------')
print(cancer.keys())
print('-----------------------')
print(cancer['data'].shape)
```

```python
x, y = cancer.data, cancer.target
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_treino)
x_treino = scaler.transform(x_treino)
x_teste = scaler.transform(x_teste)

from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(solver='adam', alpha=0.0001, hidden_layer_sizes=(10, 10, 10), random_state=42, 
                    learning_rate='constant', learning_rate_init=0.005, max_iter=200, activation='logistic', 
                    momentum=0.9, verbose=False, tol=0.0001)

mlp.fit(x_treino, y_treino)
```

```python
saida = mlp.predict(x_teste)
saida
```

```python
mlp.score(x_teste, y_teste)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_teste, saida))
print('----------------------------------------')
print(classification_report(y_teste, saida))
```
