```python
df2 = pydataset.data('iris')
```

```python
df2.head()
```

```python
plt.scatter(df2['Sepal.Length'], df2['Sepal.Width'], sizes=df2['Petal.Length'])
```

```python
a = df2.loc[df2['Species'] == 'setosa']
b = df2.loc[df2['Species'] == 'versicolor']
c = df2.loc[df2['Species'] == 'virginica']

plt.scatter(a['Sepal.Length'], a['Sepal.Width'], sizes=a['Petal.Length'], color='r')
plt.scatter(b['Sepal.Length'], b['Sepal.Width'], sizes=b['Petal.Length'], color='g')
plt.scatter(c['Sepal.Length'], c['Sepal.Width'], sizes=c['Petal.Length'], color='b')
```

```python
def specie_color(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    return 2
```

```python
df2['SpeciesNumber'] = df2['Species'].apply(specie_color)
df2.head()
```

```python
plt.scatter(df2['Sepal.Length'], df2['Sepal.Width'], sizes=20*df2['Petal.Length'], c=df2['SpeciesNumber'],cmap='viridis',alpha=0.4)
```

```python
plt.scatter(df2['Sepal.Length'], df2['Sepal.Width'], sizes=20*df2['Petal.Length'], c=df2['SpeciesNumber'],cmap='plasma',alpha=0.4)
```

```python
plt.scatter(df2['Sepal.Length'], df2['Sepal.Width'], sizes=20*df2['Petal.Length'], c=df2['SpeciesNumber'],cmap='inferno',alpha=0.4)
```

```python
plt.scatter(df2['Sepal.Length'], df2['Sepal.Width'], sizes=20*df2['Petal.Length'], c=df2['SpeciesNumber'],cmap='magma',alpha=0.4)
```
