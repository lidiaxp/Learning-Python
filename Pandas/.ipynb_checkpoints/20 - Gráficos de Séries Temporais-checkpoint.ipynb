```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import matplotlib.dates as dates
import datetime as dt
```

```python
df = pd.read_csv('ppz-jan-fev-2017.csv')
```

```python
df.head()
```

```python
def to_date(value):
    return dt.datetime(2017, 1, 1) + dt.timedelta(hours=value)  #1/1/2017 (year, month, day)
df['date'] = df['hour'].apply(to_date)
df.head()
```

```python
del df['hour']
```

```python
df.head()
```

```python
df.set_index(['date'], inplace=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot_date(df.index.to_pydatetime(), df['views'], 'b-')
ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=range(5, 32, 5))) #start, end, stepax.xaxis.set_minor_
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))  #day
ax.xaxis.grid(True, which='minor')
ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))  #month
ax.set_xlim([df.index.to_pydatetime().min(), df.index.to_pydatetime().max()])
plt.tight_layout()
plt.show()
```
