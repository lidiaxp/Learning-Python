```python
from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
```

```python
x, y = diabetes.data, diabetes.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=17, p=2)
knn.fit(x_train, y_train)
outputs = knn.predict(x_test)

print('Saída esperada: %f' % y_test[3])
print('Saída predita: %f' % outputs[3])

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, outputs)

import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.linspace(-1, 1, len(y_test)), y_test, label='data', color='black')
plt.plot(np.linspace(-1, 1, len(y_test)), outputs, label='prediction', color='red')
plt.show()
```
